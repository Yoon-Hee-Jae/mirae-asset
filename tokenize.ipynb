```python
import nltk
sentence = '''(SEOUL) - Worldcoin, an eye-scanning cryptocurrency project developed for three years by OpenAI CEO Sam Altman, creator of generative artificial intelligence (AI) ChatGPT, officially launched on Thursday, Bloomberg and Reuters reported.

The core of the Worldcoin project is the World ID, an account that only real humans can have.

First, an iris recognition device called 'Orb' is used to data an individual's iris and connect it to the blockchain, and if it is confirmed that the person is a real person, a world ID is created.

With this World ID, a cryptocurrency wallet, the World App, is created to store World Coins.

Sam Altman and Worldcoin co-founder Alex Blania have been working on Worldcoin for the past three years, recognizing that a personal digital identity is critical to distinguish humans from artificial intelligence (AI) online.

During the beta testing period alone, two million people around the world registered for a World ID. WorldCoin has been giving away free WorldCoins to encourage users to sign up.

WorldCoin has installed 1,500 orbs around the world, which it believes will allow millions more to register.

WorldCoin rallied immediately after the launch, Bloomberg reported.

Worldcoin opened at $1.70 ($2,178) per coin, rising to $3.58 ($4,587) at one point, before falling back to $2.52 ($3,229) as of 11:12 a.m. London time, according to CoinMarketCap data.

During that time, more than $145 million worth of WorldCoin was traded as exchanges like Binance listed it.

"If successful, WorldCoin will greatly increase economic opportunity and protect privacy while distinguishing humans from AI online," Oltman and Blania wrote on Twitter to coincide with the official launch of WorldCoin today.

Altman and Blania go even further, saying they believe Worldcoin could ultimately be used as a "universal basic income" for those who lose their jobs in the AI era.

"Like any ambitious project, it may or may not succeed, but progress is made by trying," Oltman said on Twitter.

But controversy has surrounded Worldcoin since before its launch.

For one, there are concerns that iris recognition information could be leaked and personal information stolen.

While WorldCoin claims that the iris images recognized by Orb are encrypted and deleted immediately afterward, some have argued that the hash value may be retained, which could compromise privacy.

There have also been reports of login information being stolen from some Orb owners, and World IDs being traded on the black market.

In Indonesia, Ghana, Chile, and elsewhere, there have been allegations of data being improperly stolen to sign people up.

"We've made the project more secure," Blania told Bloomberg, noting that in addition to the two-step identity verification process, the project now requires users to be geographically distant from where they registered their account. 

Translated with www.DeepL.com/Translator (free version)'''

# 토큰화
tokens = nltk.word_tokenize(sentence)
print('1', tokens)
# 품사 태깅
tagged = nltk.tag.pos_tag(tokens)
print('2',tagged)
# 특정 품사만 추출
nouns_list = [t[0] for t in tagged if t[1] == "NN"]
nouns_list
```
